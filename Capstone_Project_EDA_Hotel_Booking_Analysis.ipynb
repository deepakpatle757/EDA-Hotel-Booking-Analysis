## **2.Understanding The Variables**

Description of individual Variable
  
**The columns and the data it represents are listed below:**

1. **hotel :** Name of the hotel (Resort Hotel or City Hotel)

2. **is_canceled :** If the booking was canceled (1) or not (0)

3. **lead_time:** Number of days before the actual arrival of the guests

4. **arrival_date_year :** Year of arrival date

5. **arrival_date_month :** Month of month arrival date

6. **arrival_date_week_number :** Week number of year for arrival date

7. **arrival_date_day_of_month :** Day of arrival date

8. **stays_in_weekend_nights :** Number of weekend nights (Saturday or Sunday) spent at the hotel by the guests.

9. **stays_in_week_nights :** Number of weeknights (Monday to Friday) spent at the hotel by the guests.

10. **adults :** Number of adults among guests

11. **children :** Number of children among guests

12. **babies :** Number of babies among guests

13. **meal :** Type of meal booked

14. **country :** Country of guests

15. **market_segment :** Designation of market segment

16. **distribution_channel :** Name of booking distribution channel

17. **is_repeated_guest :** If the booking was from a repeated guest (1) or not (0)

18. **previous_cancellations :** Number of previous bookings that were cancelled by the customer prior to the current booking

19. **previous_bookings_not_canceled :** Number of previous bookings not cancelled by the customer prior to the current booking

20. **reserved_room_type :** Code of room type reserved

21. **assigned_room_type :** Code of room type assigned

22. **booking_changes :** Number of changes/amendments made to the booking

23. **deposit_type :** Type of the deposit made by the guest

24. **agent :** ID of travel agent who made the booking

25. **company :** ID of the company that made the booking

26. **days_in_waiting_list :** Number of days the booking was in the waiting list

27. **customer_type :** Type of customer, assuming one of four categories

28. **adr :** Average Daily Rate, as defined by dividing the sum of all lodging transactions by the total number of staying nights

29. **required_car_parking_spaces :** Number of car parking spaces required by the customer

30. **total_of_special_requests :** Number of special requests made by the customer

31. **reservation_status :** Reservation status (Canceled, Check-Out or No-Show)

32. **reservation_status_date :** Date at which the last reservation status was updated

## **3. Cleaning the Dataset**

### **Handling Duplicates Values**

```python
df[df.duplicated()]
#from this we found our that total of 119390 rows 31994 rows are duplicated row
```

*for that we remove the duplicate by following codes*

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
#now we reset the index
df.reset_index()
```

### **Handling the Null Values**

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
round(df.isnull().sum().sort_values(ascending=False)/len(df)*100,2)
```

*Let's take the company column*

```python
# for dataset it is better to drop the company column because its contains the maximumm amount of null values contains
df.drop(['company'],axis=1,inplace=True)
```

```python
df.head(10)
```

*Now,Let's take for agent column*

```python
df['agent'].mode()
#from above we see that the agent column has minimumm percentage of null values so therefore we replace that null values with mode value
df['agent'].fillna(0,inplace=True)
```

```python
df['agent'].isnull().sum()
```

### *Now,lets take the Country Column*

```python
#from above we see that percentage of null values in country column is very less so we replaced the null values by Other
df['country'].fillna(value='others',inplace=True)
```

```python
df['country'].isnull().sum()
```

### *Now let's take the children column*

```python
#from above we see that children column has very less null vaues therefore fill null values with 0
df['children'].fillna(value=0,inplace=True)
```

```python
#now check again all columns null values
df.isnull().sum()
```

```python
df.head(10)
```

```python
df.columns
```

```python
df.info()
```

### **Change the dtype of columns**

```python
#from above info it is seen that the children column and agent column has float value which is converted into int64 for our requirnment
df[['children','agent']] = df[['children','agent']].astype('int64')
```

### **Creating the new columns For Our Reqirement**

```python
#create the total no.of. guest

df['Total_guest'] = df['adults'] + df['children'] + df['babies']
```

```python
df.head(10)
```

```python
#create a column for total stay in nights
df['Total_stay_in_night'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
df.head()
```

```python
#create a column for revenue of hotel

df['revenue'] = df['Total_stay_in_night'] * df['adr']
```

```python
df.head(10)
```

### **Converting the column as our requirement**

```python
df['is_canceled'].nunique()
```

```python
#for our understanding replace the is_calceled column value of(0,1) with 'not canceled' and 'canceled'
df['is_canceled'] = df['is_canceled'].replace([0,1],['not canceled','is canceled'])
```

```python
#for our understamding replace the is_repeated_guest column of value (0,1) with the 'repeat' and 'not repeat'
df['is_repeated_guest'] = df['is_repeated_guest'].replace([0,1], ['not repeated','repeated'])
```
